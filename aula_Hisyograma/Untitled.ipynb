## Para fazer a importação da base de dados usarem a biblioteca pandas ##

```python
import pandas as pd
```

## Guardaremos o banco de dados em uma váriavel nomeada df ##

```python
df = pd.read_csv('bicicletas_salvo.csv')
```

## fazer a leitura do banco ##

```python
df.head(10)
```

## Função df.shape retorna uma tupla que indica o número de linhas e colunas ##

```python
df.shape
```

## A função df.info() fornece um resumo sobre os dados ##

```python
df.info()
```

## A função .isnull verifica os dados nulos desse data frame ##

```python
df.isnull().sum()
```

## Para tratar dados nulos, ou usamos dropna() ##

## Codigo para Interpolação Linear ##

```python
df['temperatura'] = df['temperatura'].interpolate(method='linear')
df['sensacao_termica'] = df['sensacao_termica'].interpolate(method='linear')
```

```python
df.isnull().sum()
```

## Identificar e removendo os dados duplicados ## 

```python
df.duplicated().sum()
```

## Indentificando as linhas duplicadas ##

```python
duplicadas = df.duplicated(keep=False)
```

## Selecionando as linhas duplicadas ##

```python
linhas_duplicadas = df[duplicadas]
```

## Ordenandoas linhas duplicadas pela coluna data_hora ##

```python
linhas_duplicadas.sort_values(by='data_hora')
```

```python
df_limpo = df.drop_duplicates()
```

```python
df_limpo.shape
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.displot(df_limpo, x='sensacao_termica', bins = 10)
plt.show()
```

```python
sns.displot(df_limpo, x='temperatura', bins = 10)
plt.show()
```

```python
sns.displot(df_limpo, x='umidade', bins = 10)
plt.show()
```

## O Gráfico de dispersão é útil para entendermos a relação entre as variaveis 

```python
fig, axes= plt.subplots(2,2, figsize = (15,10))

sns.scatterplot(ax= axes[0,0], data = df_limpo, x='temperatura', y='contagem', color='red')
axes[0,0].set_title('Temperatura vs Contagem')

sns.scatterplot(ax= axes[0,1], data = df_limpo, x='sensacao_termica', y='contagem', color='blue')
axes[0,1].set_title('Sensação Termica vs Contagem')

sns.scatterplot(ax= axes[1,0], data = df_limpo, x='umidade', y='contagem', color='purple')
axes[1,0].set_title('Umidade vs Contagem')

sns.scatterplot(ax= axes[1,1], data = df_limpo, x='velocidade_vento', y='contagem', color='green')
axes[1,1].set_title('Velocidade do Vento vs Contagem')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.heatmap(df_limpo.corr(numeric_only=True), annot = True, cmap='BuPu')
plt.show
```
