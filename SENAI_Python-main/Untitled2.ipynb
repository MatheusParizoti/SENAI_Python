```python
import pandas as pd

df = pd.read_csv("ATLAS.csv", encoding = 'ISO-8859-1', sep=None, engine= 'python')
df.head()
```

```python
print(df.columns)
```

```python
colunas_desejadas = ['Nome_Municipio','Sigla_UF','Data_Evento','descricao_tipologia','grupo_de_desastre','DH_MORTOS','DH_ENFERMOS','DH_DESABRIGADOS','DH_DESAPARECIDOS','PEPR_total_privado','PEPR_Indústria (R$)','PEPL_total_publico']
```

```python
len(colunas_desejadas)
```

```python
df_reduzido = df[colunas_desejadas]
df_reduzido.head()
```

```python
df_reduzido.loc[:, 'Data_Evento'] = pd.to_datetime(df_reduzido['Data_Evento'], format = '%d/%m/%Y')
```

```python
df_reduzido = df_reduzido[(df_reduzido['Data_Evento'].dt.year >= 2000) & (df_reduzido['Data_Evento'].dt.year <= 2023)]
```

```python
df_reduzido.head()
```

```python
df_reduzido.to_excel('df_reduzido.xlsx', index=False)
```

```python
import matplotlib.pyplot as plt
import seaborn  as sns

```

**Contagem de Tipos de Desastres**

```python
tipos_desastres = df_reduzido['descricao_tipologia'].value_counts()
print(tipos_desastres)
```

**Grafico de Barras Para tipo de Desastres**

```python
plt.figure(figsize=(10,6))
sns.countplot(y='descricao_tipologia',data = df_reduzido, order = tipos_desastres.index)
plt.title('distribuição dos tipos de desastres')
plt.xlabel('Contagem')
plt.ylabel('Tipos de Desastres')
plt.show()
```

**Soma dos mortos,enfermos,desabrigados e tudo mais**

```python
total_mortos= df_reduzido['DH_MORTOS'].sum()
total_enfermos= df_reduzido['DH_ENFERMOS'].sum()
total_desabrigados= df_reduzido['DH_DESABRIGADOS'].sum()
total_desaparecidos= df_reduzido['DH_DESAPARECIDOS'].sum()

print(f'Mortos: {total_mortos}')
print(f'Enfermos: {total_enfermos}')
print(f'Desabrigados: {total_desabrigados}')
print(f'Desaparecidos: {total_desaparecidos}')
```

**Grafico de Barras para os Totais**

```python
totais = [total_mortos,total_enfermos,total_desabrigados,total_desaparecidos]
categorias = ['Mortos','Enfermos','Desabrigados','Desaparecidos']

plt.figure(figsize=(8,5))
sns.barplot(x=categorias, y=totais)
plt.title('totais no geral')
plt.ylabel('Contagem')
plt.show()
```

```python
df_reduzido['PEPR_total_privado'] = df_reduzido['PEPR_total_privado'].replace('[\\$,.]','',regex=True).astype(float)
df_reduzido['PEPL_total_publico'] = df_reduzido['PEPL_total_publico'].replace('[\\$,.]','',regex=True).astype(float)
```

```python
total_privado = df_reduzido['PEPR_total_privado'].sum()
total_publico = df_reduzido['PEPL_total_publico'].sum()
```

```python
print(f'Total Privado: {total_privado}')
print(f'Total Publico: {total_publico}')
```

```python
custos = [total_privado,total_publico]
categorias_cursos = ['Privado','Público']

plt.figure(figsize=(8,5))
sns.barplot(x=categorias_cursos, y=custos)
plt.title('Os custos Totais Privados e Públicos')
plt.ylabel('Contagem')
plt.show()
```

```python
df_reduzido['Data_Evento']= pd.to_datetime(df_reduzido['Data_Evento'])

# Contagem de desastres por mês
df_reduzido['Mes'] = df_reduzido['Data_Evento'].dt.to_period('M')
desastres_por_mes = df_reduzido['Mes'].value_counts().sort_index()

# Gráfico de linha para desastres por mês
plt.figure(figsize=(12,6))
desastres_por_mes.plot(kind='line')
plt.title('Desastres')
plt.xlabel('Mês')
plt.ylabel('Numero de Desastres')
plt.show()
```

**Contagem de Desastre por Estado**

```python
desastres_por_estado = df_reduzido['Sigla_UF'].value_counts()
print(desastres_por_estado)
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x=desastres_por_estado.index, y=desastres_por_estado.values)
plt.title('Desastres por Estado')
plt.xlabel('Estado')
plt.ylabel('Numero de Desastres')
plt.show()
```

```python
#Contagem com os maiores desastres
desastres_por_municipio = df_reduzido['Nome_Municipio'].value_counts().head(10)


plt.figure(figsize=(10,6))
sns.barplot(x=desastres_por_municipio.index, y=desastres_por_municipio.values)
plt.title('TOP 10 MUNICIPIOS MAIS AFETADOS PELOS DESASTRES')
plt.xlabel('MUNICIPIO')
plt.ylabel('Numero de Desastres')
plt.xticks(rotation = 45)
plt.show()
```

**Encontrar os desastres mais comuns por estado**

```python
tabela_cruzada = pd.crosstab(df_reduzido['Sigla_UF'],df_reduzido['descricao_tipologia'])

desastre_mais_comum = tabela_cruzada.idxmax(axis=1)
print(desastre_mais_comum)
```

```python
total_ocorrencias_por_estado = tabela_cruzada.sum(axis=1)
print(total_ocorrencias_por_estado.sort_values(ascending=False))
```
