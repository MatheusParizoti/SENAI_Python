```python
df_pokemon.columns = ['Numero', 'Nome', 'Type_1', 'Type_2', 'Total', 'HP', 'Attack',
       'Defense', 'Sp_Atk', 'Sp_Def', 'Speed', 'Generation', 'isLegendary',
       'Color', 'hasGender', 'Pr_Male', 'Egg_Group_1', 'Egg_Group_2',
       'hasMegaEvolution', 'Height_m', 'Weight_kg', 'Catch_Rate',
       'Body_Style']
df_pk = df_pokemon.rename(columns={"Type_1":"Tipo_1"})
df_pk.head
```

## Leitura das colunas

```python
df_pk['Tipo_1'].head()
df_pk[['Tipo_1','Type_2']].head()
df_pk[['Tipo_1','Type_2']][10:15 + 1] ## pegar linha 10 entre 15

```

```python
colunas_desejadas = ['Numero', 'Nome', 'Type_1', 'Type_2', 'Total', 'HP', 'Attack',
       'Defense', 'Sp_Atk', 'Sp_Def', 'Speed', 'Generation', 'isLegendary']


df_pk2 = df_pokemon[colunas_desejadas]
df_pk2.head()
```

```python

df_pk2 = df_pk2.rename(columns={'Type_1':'Tipo_1',
                                          'Type_2':'Tipo_2',
                                          'HP':'Vida',
                                          'Attack':'Ataque',
                                          'Defense':'Defesa',
                                          'Sp_Atk':"Ataque_Especial",
                                          "Sp_Def":"Defesa_Especial",
                                          "Speed": "Velocidade",
                                          "Generation":"Geracao",
                                          "isLegendary":"lendario"}) #serve para renomear o nome da coluna em especifo
df_pk2.head()
```

```python
df_pk2.loc[0]
```

```python
df_pk2.loc[[0,1,17]] # linhas especificas
df_pk2.loc[0:7] # range de linhas
df_pk2.loc[0:7,['Ataque','Defesa']] # range de linhas
```

```python
# utilizando .loc realizar o filtro das linhas em que a coluna "Tipo_1" é igual
# a Grass

df_grass = df_pk2.loc[df_pk2['Tipo_1'] == 'Grass']

print(df_grass)

print(len(df_grass))
```
